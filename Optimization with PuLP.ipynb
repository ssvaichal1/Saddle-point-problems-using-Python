```python
# import PuLP
from pulp import *

# Create the 'prob' variable to contain the problem data
prob = LpProblem("Saddle Point calculation", LpMinimize)

# Create problem variables
y1=LpVariable("y1",0,None,cat = 'Float')
y2=LpVariable("y2",0, None, cat = 'Float')
y3=LpVariable("y3",0, None, cat = 'Float')
y4=LpVariable("y4",0, None, cat = 'Float')

v=LpVariable("value of the game q1",0,None,cat = 'Float')

# The objective function is added to 'prob' first
prob += v, "value of the game; to be minimized"
# The two constraints are entered
prob += 4*y1 + 3*y2 + y3 + 4*y4 <= v, "c1"
prob += 2*y1 + 5*y2 + 6*y3 + 3*y4 <= v, "c2"
prob += 1*y1 + 0*y2 + 7*y3 + 0*y4 <= v, "c3"
#prob += 2*x + y <= v, "Herb B constraint"
prob += y1+y2+y3+y4 ==1, "Sum of all must be equal to 1"
prob += y1 >= 0, "first"
prob += y2 >= 0 , "second"
prob += y3 >= 0 , "third"
prob += y4 >= 0 , "fourth"

# The problem data is written to an .lp file
prob.writeLP("Mixedq1.lp")

# The problem is solved using PuLP's choice of Solver
prob.solve()

pulp.LpStatus[prob.status]
print("y1 = {}".format(y1.varValue))
print("y2 = {}".format(y2.varValue))
print("y3 = {}".format(y3.varValue))
print("y4 = {}".format(y4.varValue))
print("value of the game = {}".format(pulp.value(prob.objective)))
```
