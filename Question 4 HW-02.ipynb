```python
import random
import numpy as np

#Randomising initial action of both players
x_ini_index = random.randint(1,3)
y_ini_index = random.randint(1,4)

A_ini = [0,0,0]
B_ini = [0,0,0,0]
n=0

A_ini[x_ini_index -1]=1
B_ini[y_ini_index -1]=1



#Creating two separate numpy matrices for history of actions and probability distribution for each player.

A_hist = np.array([A_ini])
B_hist = np.array([B_ini])


#A_pmf, B_pmf are the strategy profiles
B_pmf = B_hist
A_pmf = A_hist

U = np.array([[4,3,1,4],[2,5,6,3],[1,0,7,0]])
```

```python
for n in range(1,10000): 
    A_hist=np.append(A_hist, [[0,0,0]], axis = 0)
    A_hist[n][np.argmax(np.dot(B_pmf[n-1],U.transpose()))]=1
    
    A_pmf = np.append(A_pmf, [[A_hist[:,0].mean(),A_hist[:,1].mean(),A_hist[:,2].mean()]], axis = 0)
    
    
    #print(A_hist)
    #print("\n")
    #print(A_pmf)
    #print("\n")
    
    
    B_hist=np.append(B_hist, [[0,0,0,0]], axis = 0)
    B_hist[n][np.argmin(np.dot(A_pmf[n-1], U))]=1
    
    B_pmf = np.append(B_pmf, [[B_hist[:,0].mean(),B_hist[:,1].mean(),B_hist[:,2].mean(),B_hist[:,3].mean()]], axis = 0)
    
    
    #print(B_hist)
    #print("\n")
    #print(B_pmf)
    
    #print("#################################")
    #print("\n")
```

```python
import matplotlib.pyplot as plt
plt.plot(B_pmf[:,0], label = 'action b1')
plt.plot(B_pmf[:,1], label = 'action b2')
plt.plot(B_pmf[:,2], label = 'action b3')
plt.plot(B_pmf[:,3], label = 'action b4')
plt.grid(True)
plt.xlabel('No. of iterations')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
